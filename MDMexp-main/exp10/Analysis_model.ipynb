```python
#Age - Income - Total_Spendings - NumWebPurchases - NumStorePurchases - NumWebVisitsMonth - Recency
```

```python
features = ["Age", "Income", "Total_Spendings", "NumWebPurchases", "NumStorePurchases", "NumWebVisitsMonth", "Recency"]
```

```python
X = df[features].copy()
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
```

```python
for i in range(2, 10):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2, 10), wcss, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
kmeans = KMeans(n_clusters=6, random_state=42)
df["Cluster"] = kmeans.fit_predict(X_scaled)
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X_scaled, df["Cluster"])
print("Silhouette Score:", score)
```

```python
from sklearn.decomposition import PCA

pcs = PCA(n_components=2)
pca_data = pcs.fit_transform(X_scaled)
df["PCA1"], df["PCA2"] = pca_data[:, 0], pca_data[:, 1]
```
